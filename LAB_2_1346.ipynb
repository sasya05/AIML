```python
import pandas as pd
import matplotlib.pyplot as plt
people=['Kiran','Sasya','Rishitha','Pranay','Vinay','Sathvika']
age=[25,19,19,20,23,22]
height=[5,5,6,3,7,2]
weight=[55,43,45,57,59,50]
data={
    'Name':people,
    'Age':age,
    'Height':height,
    'Weight':weight

}
df=pd.DataFrame(data)
print(df)
#Scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df['Age'],df['Weight'],color='green',marker='o')
plt.title('Scatter plot of Age vs Weight')
plt.xlabel('Age(years)')
plt.ylabel('Weight(kgs)')
plt.grid()
plt.show()
#Bar Graph plot
plt.figure(figsize=(8,6))
plt.bar(df['Name'],df['Height'],color='purple')
plt.title("Bar Graph of Individuals\'Heights'")
plt.xlabel("Name",fontsize=15)
plt.ylabel("Height(cms)",fontsize=15)
plt.show()
#Histogram plot
plt.figure(figsize=(8,6))
plt.hist(df['Age'],bins=5,color='orange',edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age(years)')
plt.ylabel('Frequency')
plt.show()
```
